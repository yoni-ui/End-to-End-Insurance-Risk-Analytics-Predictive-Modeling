```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../data/MachineLearningRating_v3.txt", sep="|")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isna().sum().sort_values(ascending=False)
```

```python
df['TotalClaims'].hist(bins=50)
plt.title("Total Claims Distribution")
plt.show()
```

```python
df['Province'].value_counts().plot(kind='bar')
plt.title("Policies per Province")
plt.show()
```

```python
df['LossRatio'] = df['TotalClaims'] / df['TotalPremium']
df.groupby('Province')['LossRatio'].mean().sort_values()
```

```python
sns.heatmap(df[['TotalClaims','TotalPremium','SumInsured']].corr(), annot=True)
```

```python
sns.scatterplot(data=df, x="TotalPremium", y="TotalClaims", hue="PostalCode")
```
